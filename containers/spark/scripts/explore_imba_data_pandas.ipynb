```python
def read_csv_df(file_path: str):
    df = pd.read_csv(file_path)
    return df
```

```python
file_name = "aisles"
file_path = f"data/explore_data/imba_data/{file_name}/{file_name}.csv"
df = read_csv_df(file_path)
print(df.columns)
print()
print("=====Number of Nulls======")
print(df.isnull().sum())
print()
print(len(df))
df.head()
```

```python
file_name = "order_products"
path = f"data/explore_data/imba_data/{file_name}/*.csv"
csv_files = glob.glob(path)
dfs = [read_csv_df(file) for file in csv_files]
combined_df = pd.concat(dfs, ignore_index=True)
print(combined_df.columns)
print()
print("=====Number of Nulls======")
print(combined_df.isnull().sum())
print()
print(len(combined_df))
combined_df.head()
```
